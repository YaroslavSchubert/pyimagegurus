````python
import matplotlib
import skimage
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from collections import OrderedDict
import skimage as ski
from skimage.morphology import closing, opening, black_tophat
%matplotlib inline
import imutils

from __future__ import print_function
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
import os


import cv2


#HELPERS

def imshow_bgr(im):
    plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
    
    
def read_image_as_gray(path):
    image = cv2.imread(path)
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), image



def contours_on_image(image):
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    cnts, _ = cv2.findContours(gray.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

    clone = image.copy()
    
    for i, c in enumerate(cnts):
        cv2.drawContours(clone, [c], -1, (0, 255, 0), 2)
        M = cv2.moments(c)
        cX = int(M['m10'] / M['m00'])
        cY = int(M['m01'] / M['m00'])
        cv2.circle(clone, (cX, cY), 10, (0, 0, 234), -1)
        cv2.putText(clone, str(i+1), (cX - 75, cY - 25), cv2.FONT_HERSHEY_SIMPLEX, 1.25, (255, 255, 255), 4)

        
    plt.figure(figsize=(12,12))
    plt.imshow(cv2.cvtColor(clone, cv2.COLOR_BGR2RGB))
    
    return clone, cnts


        





all_images = os.listdir('data')
all_images_dict = {image:'data/'+image for image in all_images}
````

5. Rotated bounding boxes

````python
# CENTER OF MASS

image_path = 'data/more_shapes.png'
gray, image = read_image_as_gray(image_path)
(cnts, _) = cv2.findContours(gray.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
clone = image.copy()

for c in cnts:
    M = cv2.moments(c)

    cX = int(M["m10"] / M["m00"])
    cY = int(M["m01"] / M["m00"])
    cv2.circle(clone, (cX, cY), 10, (0, 255, 0), -1)


imshow_bgr(clone)


````

````python
# PERIMETER AND AREA

for (i, c) in enumerate(cnts):
    area = cv2.contourArea(c)
    perimeter = cv2.arcLength(c, True)
    print("Contour #%d -- area: %.2f, perimeter: %.2f" % (i + 1, area, perimeter))
    cv2.drawContours(clone, [c], -1, (0, 255, 0), 2)
    M = cv2.moments(c)
    cX = int(M['m10'] / M['m00'])
    cY = int(M['m01'] / M['m00'])
    cv2.putText(clone, "#%d" %(i + 1), (cX - 90, cY), cv2.FONT_HERSHEY_SIMPLEX, 1.25, (255, 255, 255), 4)

imshow_bgr(clone)
````

````python
# BOUNDING BOXES

clone = image.copy()

for c in cnts:
    (x, y, w, h) = cv2.boundingRect(c)
    cv2.rectangle(clone, (x, y), (x+w, y+h), (0, 255, 0), 2) #img, pt1, pt2, color[, thickness[, lineType[, shift]]]

imshow_bgr(clone)
````

````python
# ROTATET MIN BOUND BOX

clone = image.copy()
for c in cnts:
    box = cv2.minAreaRect(c)
    box = np.int0(cv2.cv.BoxPoints(box))
    cv2.drawContours(clone, [box], -1, (0, 255, 0), 2)

imshow_bgr(clone)
````

````python
# MIMIMUM ENCLOSING CIRCLES

clone = image.copy()
for c in cnts:
    ((x, y), radius) = cv2.minEnclosingCircle(c)
    cv2.circle(clone, (int(x), int(y)), int(radius), (0, 255, 0), 2)

imshow_bgr(clone)
````

````python
# ELLIPSE

clone = image.copy()
for c in cnts:
    if len(c) >= 5:
        ellipse = cv2.fitEllipse(c)
        cv2.ellipse(clone, ellipse, (0, 255, 0), 2)

imshow_bgr(clone)
````

````python
image = cv2.imread('data/contours_tictactoe.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cnts, _ = cv2.findContours(gray.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
for i, c in enumerate(cnts):
    area = cv2.contourArea(c)
    (x, y, w, h) = cv2.boundingRect(c)
    hull = cv2.convexHull(c)
    hullArea = cv2.contourArea(hull)
    solidity = area / float(hullArea)
    print (i,'@', solidity)
    
    char = "?"
    
    if solidity > 0.9:
        char = 'O'
    elif solidity > 0.5:
        char = "X"
    
    if char != "?":
        cv2.drawContours(image, [c], -1, (0, 255, 0), 3)
        cv2.putText(
            image, 
            char, (x, y - 10),  #string and where
            cv2.FONT_HERSHEY_SIMPLEX, 1.25, #font and how big
            (0, 255, 0), 4 #color
        )
    
imshow_bgr(image)
    
````

````python
# TETRIS BLOCKS

image = cv2.imread('data/contours_tetris_blocks.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)[1]

plt.imshow(gray,cmap='gray')
plt.imshow(thresh)

cnts, _ = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
hullImage = np.zeros(gray.shape[:2], dtype='uint8')

for (i, c) in enumerate(cnts):
    area = cv2.contourArea(c)
    (x, y, w, h) = cv2.boundingRect(c)
    
    aspectRatio = w / float(h)
    extent = area / float(w * h)
    hull = cv2.convexHull(c)
    hullArea = cv2.contourArea(hull)
    solidity = area / float(hullArea)
    
    cv2.drawContours(hullImage, [hull], -1, 255, -1)
    cv2.drawContours(image, [c], -1, (240, 0, 159), 3)
    
    if aspectRatio >= 0.98 and aspectRatio <= 1.02:
        shape = "SQUARE"
    
    elif aspectRatio >= 3.0:
        shape = "RECTANGLE"
        
    elif extent < 0.65:
        shape = "L-PIECE"
        
    elif solidity > 0.80:
        shape = "Z-PIECE"
        
    cv2.putText(
        image, 
        shape+ '  ' + str(i+1) , 
        (x, y - 10), 
        cv2.FONT_HERSHEY_SIMPLEX,
        0.5,
        (240, 0, 159), 
        2
    )
        
    print("Contour #{:d} -- aspect_ratio={:.2f}, extent={:.2f}, solidity={:.2f}".format(
        i+1, aspectRatio, extent, solidity
    ))
    


    
    
plt.figure()    
plt.imshow(hullImage, cmap='gray')

plt.figure()
plt.imshow(image)



````

# 1.11.4: Contour approximation


````python
# import the necessary packages
import cv2
 
# load the the cirles and squares image and convert it to grayscale
image = cv2.imread("data/contours_circles_and_squares.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

(cnts, _) = cv2.findContours(gray.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for c in cnts:
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.01 * peri, True)
    print(peri)
    print('---------------')
    print(approx)
    
    if len (approx) == 4:
        cv2.drawContours(image, [c], -1, (0, 255, 255), 2)
        (x, y, w, h) = cv2.boundingRect(approx)
        cv2.putText(
            image, "Rectangle", (x, y - 10), 
            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 2
        )
        
plt.figure(figsize=(12,12))
imshow_bgr(image)
````
