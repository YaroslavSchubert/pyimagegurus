````python
import matplotlib
import skimage
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from collections import OrderedDict
import skimage as ski
from skimage.morphology import closing, opening, black_tophat
%matplotlib inline
import imutils

from __future__ import print_function
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
import os


import cv2


#HELPERS

def imshow_bgr(im):
    plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
    
    
def read_image_as_gray(path):
    image = cv2.imread(path)
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), image



all_images = os.listdir('data')
all_images_dict = {image:'data/'+image for image in all_images}
````

# 1.11.1 CONTOURS

````python
image = cv2.imread('data/basic_shapes.png')
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(image, cmap='gray')
(cnts, _) = cv2.findContours(gray.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
print("number of shapes detected")
print(len(cnts))

clone = image.copy()
cv2.drawContours(clone, cnts, -1, (0, 255, 0), 2)

imshow_bgr(clone)
````

````python
canvas = np.zeros(image.shape, dtype='uint8')
# cnts[0]
number_of_countours_to_draw = -1 #all
cv2.drawContours(canvas, cnts, number_of_countours_to_draw, (0, 255, 0), 2)
plt.imshow(canvas)
````

````python
clone = image.copy()
for (i, c) in enumerate(cnts):
    cv2.drawContours(clone, [c], -1, (0, 255, 0), 2) #image, list of contours,  index[in cnts], color thickness.
    plt.subplot(2,2, i+1)
    imshow_bgr(clone)
    plt.title("contour{0!s}".format(i))
    
````

````python
clone = image.copy()
for c in cnts:
    mask = np.zeros(gray.shape, dtype='uint8')
    cv2.drawContours(mask, [c], -1, 255, -1)
    plt.figure()
    plt.imshow(image)
    plt.figure()
    plt.imshow(mask)
    plt.figure()
    plt.title('Image+MasK')
    imshow_bgr(cv2.bitwise_and(image, image, mask=mask))
````

# 1.11.2: Simple contour properties

```python
Basic properties  : 

1. Center of Mass

        M = cv2.moments(c)  
        cX = int(M["m10"] / M["m00"])  
        cY = int(M["m01"] / M["m00"]) 

        

2. Area

        area = cv2.contourArea(c)


3. Perimeter

        perimeter = cv2.arcLength(c, True)



4. Bounding boxes

        (x, y, w, h) = cv2.boundingRect(c)



5. Rotated bounding boxes

    
        box = cv2.minAreaRect(c)
        box = np.int0(cv2.cv.BoxPoints(box))
        cv2.drawContours(clone, [box], -1, (0, 255, 0), 2)



6. Minimum enclosing circles


        ((x, y), radius) = cv2.minEnclosingCircle(c)



7. Fitting an ellipse


        if len(c) >= 5:
        ellipse = cv2.fitEllipse(c)
        
```

````python
# CENTER OF MASS

image_path = 'data/more_shapes.png'
gray, image = read_image_as_gray(image_path)
(cnts, _) = cv2.findContours(gray.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
clone = image.copy()

for c in cnts:
    M = cv2.moments(c)

    cX = int(M["m10"] / M["m00"])
    cY = int(M["m01"] / M["m00"])
    cv2.circle(clone, (cX, cY), 10, (0, 255, 0), -1)


imshow_bgr(clone)


````

````python
# PERIMETER AND AREA

for (i, c) in enumerate(cnts):
    area = cv2.contourArea(c)
    perimeter = cv2.arcLength(c, True)
    print("Contour #%d -- area: %.2f, perimeter: %.2f" % (i + 1, area, perimeter))
    cv2.drawContours(clone, [c], -1, (0, 255, 0), 2)
    M = cv2.moments(c)
    cX = int(M['m10'] / M['m00'])
    cY = int(M['m01'] / M['m00'])
    cv2.putText(clone, "#%d" %(i + 1), (cX - 90, cY), cv2.FONT_HERSHEY_SIMPLEX, 1.25, (255, 255, 255), 4)

imshow_bgr(clone)
````

````python
# BOUNDING BOXES

clone = image.copy()

for c in cnts:
    (x, y, w, h) = cv2.boundingRect(c)
    cv2.rectangle(clone, (x, y), (x+w, y+h), (0, 255, 0), 2) #img, pt1, pt2, color[, thickness[, lineType[, shift]]]

imshow_bgr(clone)
````

````python
# ROTATET MIN BOUND BOX

clone = image.copy()
for c in cnts:
    box = cv2.minAreaRect(c)
    box = np.int0(cv2.cv.BoxPoints(box))
    cv2.drawContours(clone, [box], -1, (0, 255, 0), 2)

imshow_bgr(clone)
````

````python
# MIMIMUM ENCLOSING CIRCLES

clone = image.copy()
for c in cnts:
    ((x, y), radius) = cv2.minEnclosingCircle(c)
    cv2.circle(clone, (int(x), int(y)), int(radius), (0, 255, 0), 2)

imshow_bgr(clone)
````

````python
# ELLIPSE

clone = image.copy()
for c in cnts:
    if len(c) >= 5:
        ellipse = cv2.fitEllipse(c)
        cv2.ellipse(clone, ellipse, (0, 255, 0), 2)

imshow_bgr(clone)
````
